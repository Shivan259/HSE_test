```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
df = pd.read_csv('HSoE_p1_out.csv')
```

```python
ax = df.plot.scatter(x='x', y='y', c='color')
```

```python
areas = list(df['area'].unique())
```

```python
df_ars = []
for val in areas:
    df_ars.append(df.loc[df.area == val].reset_index(drop=True))
```

```python
df_ars[0]
```

```python
breaks = 0
def linebreaker(s):
    global breaks
    if len(s)<15:
        breaks = 0
        return s
    else:
        breaks = 0
        lst = s.split()
        s1 = ''
        l = 0
        for i in range(len(lst)): 
            if len(lst[i]) + l> 15:
                s1 += '\n'
                breaks += 1
                l = 0
            s1 += lst[i]
            l += len(lst[i]) + 1
            s1 += ' '
        return s1
    
print(linebreaker('aaasdaa asdfdsafgdasgjsdklag saas dd oo o o w rt y y y asdfdsafdsa asaas aa bb dafasd'))
```

```python
for k in range(len(df_ars)):
    fig = plt.figure(figsize=(15, 15), dpi=200)
    ax = fig.add_subplot()
    for i in range(4):
        a = df_ars[k][df_ars[k]['cluster'].astype(int) == i]
        if len(a)>0:
            col = a.color.iloc[0]
            lab = a.cluster_name.iloc[0]
            ax.scatter(x=a['x'], y=a['y'], c=col, label=lab, edgecolors='black')
            ax.add_artist(ax.legend(bbox_to_anchor=(1.12, 0.5)))
    for i in range(len(df_ars[k]['x'])):    
        ax.annotate(linebreaker(df_ars[k]['keyword'][i]), (df_ars[k]['x'][i], df_ars[k]['y'][i]), xytext=(-25, -12*(breaks + 1)), textcoords='offset points')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('area: ' + df_ars[k].area.iloc[0], y=-0.1, fontdict = {'fontsize': 20})
    filename = 'fig_' + str(k) +'.png'
    plt.savefig(filename, format='png')
```
