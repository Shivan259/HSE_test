```python
data = data.drop(['good (1)'], axis=1)
data.head()
```

```python
data = data.drop([49, 217])
data.sort_values(by=['count'])
```

```python
data['count'] = pd.to_numeric(data['count'])
```

```python
data.sort_values(by='y').head(25)
```

```python
data = data.drop([99])
data.sort_values(by='y').head(25)
```

```python
data['y'] = pd.to_numeric(data['y'])
```

```python
data = data[~data.duplicated(subset=['area', 'keyword'])]
```

```python
data
```

```python
data = data.reset_index(drop=True)
```

```python
data
```

```python
area_num = data.area.drop_duplicates().reset_index().drop(['index'], axis=1).reset_index().rename(columns={
    'area': 'area',
    'index': 'area_num'})
area_num
```

```python
data = data.merge(area_num, on='area')
data.head()
```

```python
pallete = ['#202020', '#FF1010', '#10FF10', '#1010FF', '#FFFF10',
           '#FF10FF', '#10FFFF', '#808020', '#802080', '#208080',
           '#AA5050', '#50AA50', '#5050AA', '#FF5000', '#0080FF']
sns.palplot(pallete)
```

```python
colors = []
for i in range(217):
    colors.append(pallete[int(data.cluster[i] + data.area_num[i]) % 15])
```

```python
data["color"] = colors
```

```python
data.head()
```

```python
data = data.drop(['area_num'], axis=1)
data.head()
```

```python
data.sort_values(by=['count'], ascending=False).sort_values(by=['area', 'cluster', 'cluster_name'])
```

```python
data.sort_values(by=['count'], ascending=False).sort_values(by=['area', 'cluster', 'cluster_name']).reset_index(drop=True)
```

```python
data = data.sort_values(by=['count'], ascending=False).sort_values(by=['area', 'cluster', 'cluster_name']).reset_index(drop=True)
```

```python
data.to_csv('HSoE_p1_out.csv', index=False)
```
